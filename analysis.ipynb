# Analysis max_return.csv

```python
import pandas as pd
```

```python
max_return_csv = "results/max_return.csv"

df = pd.read_csv(max_return_csv, sep=";")
df
```

```python
agg_df = df.groupby(["model", "env", "env_args"]).agg(
    count = ("seed", "count"),
    max_return_mean = ("max_return", "mean"),
    max_return_std  = ("max_return", "std"),
    test_max_return_mean = ("test_max_return", "mean"),
    test_max_return_std  = ("test_max_return", "std"),
)
agg_df
```
