# Analysis max_return.csv

```python
import pandas as pd
import numpy as np
import scipy.stats as st
```

```python
max_return_csv = "results/return.csv"

df = pd.read_csv(max_return_csv, sep=";")
try:
    def func(x):
        if type(x) == str:
            return np.array(list(map(float, x.strip('[] ,').split(', '))))
    df["test_return"] = df["test_return"].map(func)
except ValueError as e:
    print(e)
df
```

```python
df['test_return_mean_run'] = df['test_return'].apply(np.mean)
df['test_return_std_run'] = df['test_return'].apply(np.std)
df
```

```python
agg_per_seed_df = df.groupby(["env", "env_args", "model"]).agg(
    count = ("seed", "count"),
	concat = ("test_return_mean_run", lambda x: list(x)),
    test_return_mean = ("test_return_mean_run", "mean"),
    test_return_std  = ("test_return_std_run", "std"),
)
agg_per_seed_df["test_return_tex"] = (
    agg_per_seed_df["test_return_mean"].map("{:.2f}".format)
    + "$\pm$"
    + agg_per_seed_df["test_return_std"].map("{:.2f}".format)
)
agg_per_seed_df
```

```python
test_return_table = agg_per_seed_df.reset_index().pivot(index=["env", "env_args"], columns="model", values=["test_return_mean", "concat", "test_return_tex"])
test_return_table
```

```python
best_model = test_return_table["test_return_mean"].idxmax(axis=1)
best_model
```

```python
def func(row, best_model):
    best_model_name = best_model.loc[row.name]
    best_model_array = row["concat"].loc[best_model_name]
    print(f"for {row.name} the best model is {best_model_name}")
    row.loc["test_return_tex", best_model_name] += " *"
    for model, concat in row["concat"].items():
        if model != best_model_name:
            res = st.ttest_ind(best_model_array, concat, equal_var=False)
            if res.pvalue > 0.05:
                print(f"{model} is not significatively different")
                row.loc["test_return_tex", model] += " *"
    return row

test_return_table = test_return_table.apply(func, args=(best_model,), axis=1)
test_return_table["test_return_tex"]
```

```python
latex_to_print = test_return_table["test_return_tex"].to_latex(
        index=True, formatters={"name": str.upper}, float_format="{:.1f}".format
    )
print(latex_to_print)
```
